```python
#Case 1:-
# Static input values
temperature = 25.0  # Temperature in Celsius
humidity = 40.0     # Humidity in %
wind_speed = 15.0   # Wind speed in km/h

# Input validation
if not (-50 <= temperature <= 50):
    print("Error: Temperature must be between -50°C and 50°C.")
elif not (0 <= humidity <= 100):
    print("Error: Humidity must be between 0% and 100%.")
elif wind_speed < 0:
    print("Error: Wind speed must be non-negative.")
else:
    # Quadratic formula for prediction
    prediction = 0.5 * temperature**2 - 0.2 * humidity + 0.1 * wind_speed - 15

    # Determine the weather category
    if prediction > 300:
        category = "Sunny"
    elif 200 < prediction <= 300:
        category = "Cloudy"
    elif 100 < prediction <= 200:
        category = "Rainy"
    else:
        category = "Stormy"

    # Output results
    print("Static Inputs:")
    print(f"  Temperature: {temperature}°C")
    print(f"  Humidity: {humidity}%")
    print(f"  Wind Speed: {wind_speed} km/h")
    print("\nPrediction Results:")
    print(f"  Prediction Score (P): {prediction}")
    print(f"  Weather Category: {category}")
```
